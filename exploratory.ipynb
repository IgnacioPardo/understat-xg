```python
# plot shots
""" 
liv_man_city_shots.columns =>

Index(['Team', 'Minute', 'Result', 'X', 'Y', 'xG', 'Player', 'Home/Away',
       'Player ID', 'Situation', 'Season', 'Shot Type', 'Match ID',
       'Home Team', 'Away Team', 'Home Goals', 'Away Goals', 'Date',
       'Player Assisted', 'Last Action'],
      dtype='object')
"""

import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

# Set the size of the plot
fig.set_size_inches(7, 5)

# draw the pitch

# Pitch Outline & Centre Line
plt.plot([0, 0], [0, 90], color="black")
plt.plot([0, 130], [90, 90], color="black")
plt.plot([130, 130], [90, 0], color="black")
plt.plot([130, 0], [0, 0], color="black")
plt.plot([65, 65], [0, 90], color="black")

# Left Penalty Area
plt.plot([16.5, 16.5], [65, 25], color="black")
plt.plot([0, 16.5], [65, 65], color="black")
plt.plot([16.5, 0], [25, 25], color="black")

# Right Penalty Area
plt.plot([130, 113.5], [65, 65], color="black")
plt.plot([113.5, 113.5], [65, 25], color="black")
plt.plot([113.5, 130], [25, 25], color="black")

# Left 6-yard Box
plt.plot([0, 5.5], [54, 54], color="black")
plt.plot([5.5, 5.5], [54, 36], color="black")
plt.plot([5.5, 0.5], [36, 36], color="black")

# Right 6-yard Box
plt.plot([130, 124.5], [54, 54], color="black")
plt.plot([124.5, 124.5], [54, 36], color="black")
plt.plot([124.5, 130], [36, 36], color="black")

# Make Circles
centreCircle = plt.Circle((65, 45), 9.15, color="black", fill=False)
centreSpot = plt.Circle((65, 45), 0.8, color="black")

# 6 Yard box circles
leftPenSpot = plt.Circle((11, 45), 0.8, color="black")
rightPenSpot = plt.Circle((119, 45), 0.8, color="black")

ax.add_artist(centreCircle)
ax.add_artist(centreSpot)
ax.add_artist(leftPenSpot)
ax.add_artist(rightPenSpot)

# Definir colores válidos
valid_colors = {
    "Goal_home": "orange",
    "Saved_home": "darkorange",
    "Other_home": "peachpuff",
    "Goal_away": "blue",
    "Saved_away": "darkblue",
    "Other_away": "lightblue",
}

# Asignar colores basados en las condiciones
colors = np.where(
    liv_man_city_shots["Team"] == "h",
    np.where(
        liv_man_city_shots["Result"] == "Goal",
        valid_colors["Goal_home"],
        np.where(
            liv_man_city_shots["Result"] == "Saved",
            valid_colors["Saved_home"],
            valid_colors["Other_home"],
        ),
    ),
    np.where(
        liv_man_city_shots["Result"] == "Goal",
        valid_colors["Goal_away"],
        np.where(
            liv_man_city_shots["Result"] == "Saved",
            valid_colors["Saved_away"],
            valid_colors["Other_away"],
        ),
    ),
)

# Invertir las posiciones en el campo para los equipos visitantes
liv_man_city_shots["X_adj"] = np.where(
    liv_man_city_shots["Team"] == "a",
    1 - liv_man_city_shots["X"],
    liv_man_city_shots["X"],
)

liv_man_city_shots["Y_adj"] = np.where(
    liv_man_city_shots["Team"] == "a",
    1 - liv_man_city_shots["Y"],
    liv_man_city_shots["Y"],
)

# Graficar los puntos
ax.scatter(
    liv_man_city_shots["X_adj"] * 130,
    liv_man_city_shots["Y_adj"] * 90,
    s=liv_man_city_shots["xG"] * 100,
    c=colors,
)

# Acumulated xG vs actual goals

liv_acc_xG = liv_man_city_shots[liv_man_city_shots["Team"] == "h"]["xG"].cumsum()
liv_goals = liv_man_city_shots[liv_man_city_shots["Team"] == "h"][
    liv_man_city_shots["Result"] == "Goal"
].shape[0]

city_acc_x = liv_man_city_shots[liv_man_city_shots["Team"] == "a"]["xG"].cumsum()
city_goals = liv_man_city_shots[liv_man_city_shots["Team"] == "a"][
    liv_man_city_shots["Result"] == "Goal"
].shape[0]


ax.set_title(f"Liverpool vs Manchester City - Shots \n Goals: {liv_goals} - {city_goals} \n axG: {liv_acc_xG.iloc[-1]:.2f} - {city_acc_x.iloc[-1]:.2f}")
```

```python
# mplsoccer
# !pip3 install mplsoccer

from mplsoccer import Pitch, VerticalPitch

PITCH_WIDTH = 68
PITCH_HEIGHT = 105

pitch = Pitch(
    pitch_type='custom', 
    pitch_width=PITCH_WIDTH,
    pitch_length=PITCH_HEIGHT,
)
fig, ax = pitch.draw()

ax.scatter(
    liv_man_city_shots["X_adj"] * PITCH_HEIGHT,
    liv_man_city_shots["Y_adj"] * PITCH_WIDTH,
    s=liv_man_city_shots["xG"] * 100,
    c=colors,
    label=liv_man_city_shots["Team"]
)

ax.set_title(f"Liverpool vs Manchester City - Shots \n Goals: {liv_goals} - {city_goals} \n axG: {liv_acc_xG.iloc[-1]:.2f} - {city_acc_x.iloc[-1]:.2f}")
```

```python
# Liverpool side

pitch = VerticalPitch(
    pitch_type='custom', 
    pitch_width=PITCH_WIDTH,
    pitch_length=PITCH_HEIGHT,
    half=True
)

fig, ax = pitch.draw()

ax.scatter(
    1 - liv_man_city_shots["Y_adj"] * -PITCH_WIDTH ,
    1 - liv_man_city_shots["X_adj"] * -PITCH_HEIGHT,
    s=liv_man_city_shots["xG"] * 100,
    c=colors,
)

ax.set_title(f"Liverpool vs Manchester City - Shots \n Goals: {liv_goals} - {city_goals} \n axG: {liv_acc_xG.iloc[-1]:.2f} - {city_acc_x.iloc[-1]:.2f}")
```

```python
# Timing chart from shots

timed_shots_home = liv_man_city_shots[liv_man_city_shots["Team"] == "h"].sort_values("Minute")[["Minute", "xG", "Result"]]
timed_shots_away = liv_man_city_shots[liv_man_city_shots["Team"] == "a"].sort_values("Minute")[["Minute", "xG", "Result"]]

# add minute 0, cumulative xG 0
timed_shots_home = pd.concat([
    pd.DataFrame({"Minute": [0], "xG": [0], "Result": [""]}), 
    timed_shots_home], ignore_index=True
)

timed_shots_away = pd.concat([
    pd.DataFrame({"Minute": [0], "xG": [0], "Result": [""]}), 
    timed_shots_away], ignore_index=True
)

timed_shots_home["cum_xG"] = timed_shots_home["xG"].cumsum()
timed_shots_away["cum_xG"] = timed_shots_away["xG"].cumsum()

goals_home = timed_shots_home[timed_shots_home["Result"] == "Goal"]
goals_away = timed_shots_away[timed_shots_away["Result"] == "Goal"]

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(timed_shots_home["Minute"], timed_shots_home["cum_xG"], color="orange", linewidth=2, label="Liverpool")
ax.plot(timed_shots_away["Minute"], timed_shots_away["cum_xG"], color="blue", linewidth=2, label="Manchester City")

# plot point for goals
ax.scatter(goals_home["Minute"], goals_home["cum_xG"], color="orange", s=100)
ax.scatter(goals_away["Minute"], goals_away["cum_xG"], color="blue", s=100)

ax.set_title(f"Liverpool vs Manchester City - Accumulated xG over time \n Goals: {liv_goals} - {city_goals} \n axG: {liv_acc_xG.iloc[-1]:.2f} - {city_acc_x.iloc[-1]:.2f}")
ax.set_xlabel("Minute")
ax.set_ylabel("Accumulated xG")

ax.legend()

plt.show()
```

```python
def ellipse_arc(
    x_center=0,
    y_center=0,
    a=1,
    b=1,
    start_angle=0,
    end_angle=2 * np.pi,
    N=100,
    closed=False,
):
    t = np.linspace(start_angle, end_angle, N)
    x = x_center + a * np.cos(t)
    y = y_center + b * np.sin(t)
    path = f"M {x[0]}, {y[0]}"
    for k in range(1, len(t)):
        path += f"L{x[k]}, {y[k]}"
    if closed:
        path += " Z"
    return path

n_strips = 12
pitch_lines_color = "white"
grass_color = "#3FAF48"
grass_color_alt = "#31923B"
color_for_index = lambda i: grass_color if i % 2 == 0 else grass_color_alt

grass = [
    go.layout.Shape(
        layer="below",
        type="rect",
        x0=-(PITCH_HEIGHT / n_strips / 2),
        y0=-(PITCH_HEIGHT / n_strips / 2),
        x1=PITCH_HEIGHT + (PITCH_HEIGHT / n_strips / 2),
        y1=PITCH_WIDTH + (PITCH_HEIGHT / n_strips / 2),
        fillcolor=grass_color_alt,
        line=dict(color=pitch_lines_color),
        name=f"Grass_bg",
    )
] + [
    # Green strips for grass, alternating between lightgreen and green
    go.layout.Shape(
        layer="below",
        type="rect",
        x0=i * PITCH_HEIGHT / n_strips,
        y0=0,
        x1=(i + 1) * PITCH_HEIGHT / n_strips,
        y1=PITCH_WIDTH,
        fillcolor=color_for_index(i),
        line=dict(color=pitch_lines_color),
        name=f"Grass_{i}",
    )
    for i in range(0, n_strips)
]

pitch_shapes = [
    # Pitch Outline
    go.layout.Shape(
        layer="below",
        type="rect",
        x0=0,
        y0=0,
        x1=PITCH_HEIGHT,
        y1=PITCH_WIDTH,
        line=dict(color=pitch_lines_color),
    ),
    # Midline
    go.layout.Shape(
        layer="below",
        type="line",
        x0=PITCH_HEIGHT / 2,
        y0=0,
        x1=PITCH_HEIGHT / 2,
        y1=PITCH_WIDTH,
        line=dict(color=pitch_lines_color),
    ),
    # Left Penalty Area
    go.layout.Shape(
        layer="below",
        type="rect",
        x0=0,
        y0=(PITCH_WIDTH - 40.3) / 2,
        x1=16.5,
        y1=PITCH_WIDTH - (PITCH_WIDTH - 40.3) / 2,
        line=dict(color=pitch_lines_color),
    ),
    # Right Penalty Area
    go.layout.Shape(
        layer="below",
        type="rect",
        x0=PITCH_HEIGHT - 16.5,
        y0=(PITCH_WIDTH - 40.3) / 2,
        x1=PITCH_HEIGHT,
        y1=PITCH_WIDTH - (PITCH_WIDTH - 40.3) / 2,
        line=dict(color=pitch_lines_color),
    ),
    # Left 6-yard Box
    go.layout.Shape(
        layer="below",
        type="rect",
        x0=0,
        y0=(PITCH_WIDTH - 18.3) / 2,
        x1=5.5,
        y1=PITCH_WIDTH - (PITCH_WIDTH - 18.3) / 2,
        line=dict(color=pitch_lines_color),
    ),
    # Right 6-yard Box
    go.layout.Shape(
        layer="below",
        type="rect",
        x0=PITCH_HEIGHT - 5.5,
        y0=(PITCH_WIDTH - 18.3) / 2,
        x1=PITCH_HEIGHT,
        y1=PITCH_WIDTH - (PITCH_WIDTH - 18.3) / 2,
        line=dict(color=pitch_lines_color),
    ),
    # Goals
    go.layout.Shape(
        layer="below",
        type="rect",
        x0=-2,
        y0=(PITCH_WIDTH - 7.32) / 2,
        x1=0,
        y1=PITCH_WIDTH - (PITCH_WIDTH - 7.32) / 2,
        line=dict(color=pitch_lines_color),
    ),
    go.layout.Shape(
        layer="below",
        type="rect",
        x0=PITCH_HEIGHT,
        y0=(PITCH_WIDTH - 7.32) / 2,
        x1=PITCH_HEIGHT + 2,
        y1=PITCH_WIDTH - (PITCH_WIDTH - 7.32) / 2,
        line=dict(color=pitch_lines_color),
    ),
    # Penalty spots
    go.layout.Shape(
        layer="below",
        type="circle",
        x0=11,
        y0=PITCH_WIDTH / 2 - 0.8,
        x1=11.2,
        y1=PITCH_WIDTH / 2 - 0.6,
        line=dict(color=pitch_lines_color),
    ),
    go.layout.Shape(
        layer="below",
        type="circle",
        x0=PITCH_HEIGHT - 11,
        y0=PITCH_WIDTH / 2 - 0.8,
        x1=PITCH_HEIGHT - 11.2,
        y1=PITCH_WIDTH / 2 - 0.6,
        line=dict(color=pitch_lines_color),
    ),
    # Arcs
    # Left
    go.layout.Shape(
        layer="below",
        type="path",
        path=ellipse_arc(
            x_center=11,
            y_center=PITCH_WIDTH / 2,
            a=9.15,
            b=9.15,
            start_angle=-np.pi / 3.4,
            end_angle=np.pi / 3.4,
        ),
        line=dict(color=pitch_lines_color),
    ),
    # Right
    go.layout.Shape(
        layer="below",
        type="path",
        path=ellipse_arc(
            x_center=PITCH_HEIGHT - 11,
            y_center=PITCH_WIDTH / 2,
            a=9.15,
            b=9.15,
            start_angle=np.pi - np.pi / 3.4,
            end_angle=np.pi + np.pi / 3.4,
        ),
        line=dict(color=pitch_lines_color),
    ),
    # Center circle
    go.layout.Shape(
        layer="below",
        type="circle",
        x0=PITCH_HEIGHT / 2 - 9.15,
        y0=PITCH_WIDTH / 2 - 9.15,
        x1=PITCH_HEIGHT / 2 + 9.15,
        y1=PITCH_WIDTH / 2 + 9.15,
        line=dict(color=pitch_lines_color),
    ),
    # Center spot
    go.layout.Shape(
        layer="below",
        type="circle",
        x0=PITCH_HEIGHT / 2 - 0.8,
        y0=PITCH_WIDTH / 2 - 0.8,
        x1=PITCH_HEIGHT / 2 + 0.8,
        y1=PITCH_WIDTH / 2 + 0.8,
        line=dict(color=pitch_lines_color),
    ), 
]
```

```python
# Definir colores válidos
css_valid_colors = {
    "Goal_home": "orange",
    "Saved_home": "darkorange",
    "Other_home": "lightcoral",
    "Goal_away": "blue",
    "Saved_away": "darkblue",
    "Other_away": "lightblue",
}


liv_man_city_shots["plotly_color"] = np.where(
    liv_man_city_shots["Team"] == "h",
    np.where(
        liv_man_city_shots["Result"] == "Goal",
        css_valid_colors["Goal_home"],
        np.where(
            liv_man_city_shots["Result"] == "Saved",
            css_valid_colors["Saved_home"],
            css_valid_colors["Other_home"],
        ),
    ),
    np.where(
        liv_man_city_shots["Result"] == "Goal",
        css_valid_colors["Goal_away"],
        np.where(
            liv_man_city_shots["Result"] == "Saved",
            css_valid_colors["Saved_away"],
            css_valid_colors["Other_away"],
        ),
    ),
)

# Sets the marker symbol type. Adding 100 is equivalent to appending “-open” to a symbol name. Adding 200 is equivalent to appending “-dot” to a symbol name. Adding 300 is equivalent to appending “-open-dot” or “dot-open” to a symbol name.
# The ‘symbol’ property is an enumeration that may be specified as:
# One of the following enumeration values:
# [0, ‘0’, ‘circle’, 100, ‘100’, ‘circle-open’, 200, ‘200’, ‘circle-dot’, 300, ‘300’, ‘circle-open-dot’, 1, ‘1’, ‘square’, 101, ‘101’, ‘square-open’, 201, ‘201’, ‘square-dot’, 301, ‘301’, ‘square-open-dot’, 2, ‘2’, ‘diamond’, 102, ‘102’, ‘diamond-open’, 202, ‘202’, ‘diamond-dot’, 302, ‘302’, ‘diamond-open-dot...

# 'SavedShot', 'BlockedShot', 'Goal', 'MissedShots', 'ShotOnPost'

liv_man_city_shots["symbol"] = liv_man_city_shots["Result"].apply(
    lambda x: "star" if x == "Goal" else 
        "square" if x == "SavedShot" else
        "diamond" if x == "BlockedShot" else
        "circle" if x == "MissedShots" else
        "cross" if x == "ShotOnPost" else
        "circle"
)
```
